```python
import numpy as np, matplotlib.pyplot as plt, astropy.constants as const, astropy.units as u, matplotlib, pyneb as pn
from astropy.io import fits
%config InlineBackend.figure_format = 'retina'
plt.style.use('../astro406/mplstyle_cb.txt')
```

useful constants

```python
c = const.c.cgs.value   
k_B = const.k_B.cgs.value
m_p = const.m_p.cgs.value 
m_H = m_p  # mass of hydrogen atom in grams
m_O = 16 * m_p  # mass of oxygen atom in grams

O3 = pn.Atom(atom = 'O3')  # O3
H1 = pn.RecAtom('H', 1)  # H-beta
```

read in all necessary data

```python
#==================== spectral data ====================
hdul = fits.open('data/SMC_SNR/b0886.norm.dc.fits'); bflux = hdul[0].data
hdul = fits.open('data/SMC_SNR/r0886.norm.dc.fits'); rflux = hdul[0].data
if 'CD1_1' in list(hdul[0].header.keys()):
    wave = hdul[0].header['CRVAL1'] + hdul[0].header['CD1_1'] * (np.arange(hdul[0].header['NAXIS1']) - hdul[0].header['CRPIX1'] + 1)
else: wave = hdul[0].header['CRVAL1'] + hdul[0].header['CDELT1'] * (np.arange(hdul[0].header['NAXIS1']) - hdul[0].header['CRPIX1'] + 1)

# noise is the sqrt of the variance
hdul = fits.open('data/SMC_SNR/b0886_var.norm.dc.fits'); bnoise = np.sqrt(hdul[0].data)
hdul = fits.open('data/SMC_SNR/r0886_var.norm.dc.fits'); rnoise = np.sqrt(hdul[0].data)

noise = np.concatenate((bnoise, rnoise))
flux = np.concatenate((bflux, rflux))

#==================== fiber data =====================
hex = np.loadtxt('data/fiber_info/STD Sky to CCDs-Table 1.csv', delimiter=',', skiprows=17,usecols=(5,6), max_rows=552)
sorted_data = []  # sort the data in a snake-like pattern (native to IFUM)
for i, y in enumerate(np.unique(hex[:, 1])):  # get all vertical values
    row_data = hex[ hex[:, 1] == y ]  # just take information for each individual row
    sorted_indices = np.argsort(row_data[:, 0])  # sort by x coord
    if i % 2 == 0:  # even index rows go right-to-left, odd go left-to-right
        sorted_indices = sorted_indices[::-1]
    sorted_data.extend(row_data[sorted_indices])
hex = np.array(sorted_data)
```

function for plotting maps

```python
def maps(cvals, ax, fig, vlim=None, c=0):
    if vlim == None: vmin=np.percentile(cvals, 10); vmax=np.percentile(cvals, 90)
    else: vmin,vmax = vlim
    r = 0.6 # radius of hexagons
    if c==0: cmap = matplotlib.colormaps['magma']  # for intensity maps
    if c==1: cmap = matplotlib.colormaps['bwr_r']  # for velocity maps
    if c==2: cmap = matplotlib.colormaps['Blues']  
    norm = matplotlib.colors.Normalize(vmin=vmin, vmax=vmax)
    for i,x,y in zip(range(len(cvals)), hex[:,0], hex[:,1]):
        color = cmap(norm(cvals[i]))
        hexag = matplotlib.patches.RegularPolygon((x,y), numVertices=6, radius=r, facecolor=color)#, edgecolor='k', linewidth=0.25)
        ax.add_patch(hexag)
        # ax.text(x, y, f'{i}', ha='center', va='center', fontsize=6, color='red')  # takes a long time, reduce hexagons to see numbers
        
    sm = matplotlib.cm.ScalarMappable(norm=norm, cmap=cmap); fig.colorbar(sm, ax=ax, location='right')
    ax.set_xlim(np.min(hex[:,0])-r, np.max(hex[:,0])+r)
    ax.set_ylim(np.min(hex[:,1])-r, np.max(hex[:,1])+r)
    ticks = [-10,-5,0,5,10]; labels = [rf'$\mathrm{{{t}"}}$' for t in ticks]
    ax.set_xticks(ticks); ax.set_xticklabels(labels); ax.set_yticks(ticks); ax.set_yticklabels(labels)
    ax.set_aspect('equal');
```

auxiliary functions for getspectrum

```python
def getflux(ion, wl, T, ne, ni = None):
    if ni is None: ni = ne * 1e-4  # rough estimate from ISM densities
    eps = ion.getEmissivity(tem=T, den=ne, wave=wl)
    j = eps * ne * ni # emissivity per unit volume
    return j

def weights(x, y, R, dR=None):
    if dR is None: dR = 0.1 * R
    rho2 = x**2 + y**2
    Rin2, Rout2 = (R - dR)**2, (R + dR)**2
    w = np.where((rho2 >= Rin2) & (rho2 <= Rout2), 2.0 * np.sqrt(Rout2 - rho2), 0)
    return w

# line-of-sight velocities for a spherical shell as seen on IFUM
def los_vel(x, y, R, k, dR=None):
    if dR is None: dR = 0.1 * R
    rho2 = x**2 + y**2
    Rin2, Rout2 = (R - dR)**2, (R + dR)**2
    mask = (rho2 >= Rin2) & (rho2 <= Rout2)  # only include material between Rin and Rout
    v_near = np.where(mask, (-k / R**2) * np.sqrt(Rout2 - rho2), 0)
    return v_near

# thermal broadening 
def broaden(wl0, T, mass):
    v_th = np.sqrt(2 * k_B * T / mass)
    return wl0 * (v_th / c)

# gaussian profile
def getgauss(wl, wl0, sigma, flux):
    amp = flux / (np.sqrt(2 * np.pi) * sigma)
    profile = amp * np.exp(-0.5 * ((wl - wl0) / sigma)**2)
    return profile, amp

# doppler shift
def dopplershift(wl0, vel):
   ckm = c / 1e5  # speed of light in km/s
   return wl0 * (vel / ckm)
```

function to reproduce the spectra

```python
def getspectrum(wl, T, ne, lines, vel, x, y, R, idx, dR, static=False):
    spectrum = np.zeros_like(wl, dtype=float)
    vel *= 1e5  # cm/s
    for line in lines:
        ion, lam0, mass, ni = line['ion'], line['wl'], line['mass'], line.get('ni', 1e-4 * ne)
        j = getflux(ion, lam0, T, ne, ni=ni)              # emissivity per unit volume
        sigma_th = broaden(lam0, T, mass)                 # thermal broadening of lines
        sigma = np.sqrt(sigma_th**2 + ((lam0/c)*vel)**2)  # bulk velocity broadening
        if not static:
            F = j * weights(x[idx], y[idx], R, dR)                 # get observed flux from geometry
            spectrum += getgauss(wl, lam0*(1+vel/c), sigma, F)[0]  # doppler shifted lines
        else: spectrum += getgauss(wl, lam0, sigma, j)[0]          # static lines, no doppler shift and no geometry
    return spectrum
```

function to run the model given some info about the SNR

```python
def runmodel(idx, centers, Rlist, klist, dRlist=None, T=2600, ne=30, velstat=22):
    X, Y = [], []
    for i in centers: X.append(hex[:,0] - hex[:,0][i]); Y.append(hex[:,1] - hex[:,1][i])
    
    velocity=[]  # get the velocity in each spaxel for each shell
    for i, (R, k) in enumerate(zip(Rlist, klist)): velocity.append(los_vel(X[i], Y[i], R, k))
    
    lines = [
        {'ion': O3,  'wl': 4959, 'mass': m_O},
        {'ion': O3,  'wl': 5007, 'mass': m_O},
        {'ion': H1,  'wl': 4861, 'mass': m_H}]
    
    mocktotal=[]; mockspec=[]
    for i in range(len(flux)):
        shells = []
        for shell in range(len(Rlist)):
            R=Rlist[shell]; vel=velocity[shell][i]
            shells.append(getspectrum(wave, T, ne, lines[:2], vel=vel, R=R, dR=dRlist[shell], idx=i, x=X[shell], y=Y[shell]))
        spec_tot = np.sum(shells, axis=0)
        static = getspectrum(wave, T, ne, lines, vel=velstat, idx=i, R=None, dR=None, x=None, y=None, static=True)
        mockspec.append(spec_tot + static)
        mocktotal.append(np.sum(mockspec[-1]))
    
    intensity = np.array(mocktotal) / np.array(mocktotal).max()
    scale = np.max(flux[idx]) / np.max(mockspec[idx])
    spectrum = mockspec[idx] * scale 
    velocity = np.sum(velocity, axis=0)

    return spectrum, intensity, velocity
```

```python
n=203
# shells is [center, velocity, thickness]
shells = [[332, 12, 10000, 1.25], [332, 11, -5000, 1]]  
centers = [shell[0] for shell in shells]  
Rlist   = [shell[1] for shell in shells]  # radii of shells in arcsec
klist   = [shell[2] for shell in shells]  # velocity scaling factors for shells
dRlist  = [shell[3] for shell in shells]  # shell thicknesses

spectrum, intensity, velocity = runmodel(idx=n, centers=centers, Rlist=Rlist, klist=klist, dRlist=dRlist)

fig,ax=plt.subplots(1,1,layout='compressed')
ax.plot(wave/(1+(5010-5007)/5007), flux[n]-np.median(flux[n]), lw=0.75)
ax.plot(wave, spectrum, lw=0.75, ls='--')
ax.set_box_aspect(0.5)

fig, ax = plt.subplots(1,2,layout='compressed')
maps(intensity, ax[0], fig, c=0)
maps(-velocity, ax[1], fig, vlim=(-1500,1500), c=1)
ax[1].scatter(hex[n][0], hex[n][1], s=5, alpha=0.5, c='k')
ax[1].scatter([hex[center][0] for center in centers], [hex[center][1] for center in centers], s=10, alpha=0.5, c='green', marker='x')
ax[0].set_title(r'$\rm Intensity$'); ax[1].set_title(r'$\rm Velocity\ (km \cdot s^{-1})$');
```
